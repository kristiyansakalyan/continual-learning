```python
def parse_video_splits(content):
    training = []
    validation = []
    test = []
    current_list = None
    
    for line in content:
        line = line.strip()
        if not line:
            continue
        if line.startswith("# Training"):
            current_list = training
        elif line.startswith("# Validation"):
            current_list = validation
        elif line.startswith("# Test"):
            current_list = test
        elif line.startswith("Video"):
            if current_list is not None:
                current_list.append(line)
    
    return {
        "train": training,
        "val": validation,
        "test": test
    }

with open("data/CaDISv2/splits.txt", "r") as file:
    data = file.readlines()

parse_video_splits(data)
```
