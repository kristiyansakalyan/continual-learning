```python
import os
import json

annotations_dir = "data/catract-1k/Annotations/Coco-Annotations"

cases = [case for case in os.listdir(annotations_dir) if case.startswith("case_")]

CATARACT_1K_CATEGORIES: dict[str, int] = {
    "Cornea": 1,
    "Katena Forceps": 2,
    "cornea1": 3,
    "Lens Injector": 4,
    "Irrigation-Aspiration": 5,
    "Capsulorhexis Forceps": 6,
    "Spatula": 7,
    "pupil1": 8,
    "Phacoemulsification Tip": 9,
    "Incision Knife": 10,
    "Pupil": 11,
    "Slit Knife": 12,
    "Lens": 13,
    "Capsulorhexis Cystotome": 14,
    "Gauge": 15,
}

case_cat_to_idx: dict[str, dict[int, int]] = {}

for case in cases:
    
    with open(f"{annotations_dir}/{case}/annotations/instances.json", "r") as file:
        data = json.load(file)
    categories = data["categories"]
    cat_to_idx_curr = {
        entry["id"]: CATARACT_1K_CATEGORIES[entry["name"]]
        for entry in categories
    } 

    case_cat_to_idx[case] = cat_to_idx_curr

case_cat_to_idx
    
```
